```python
data['Brand']=data['Brand'].str.title()
```

```python
data['Revenue_per_unit']= data['Net_Amount']/data['Unit_Price']
```

```python
data.head(5)
```

```python
bins = [20,30,40,50,60,70]
labels=['21-30','31-40','41-50','51-60','61-70']
data['Age_Group']= pd.cut(data['Customer_Age'],bins=bins,labels=labels)
```

```python
data.head(5)
```

```python
data['Is_Weekend']=data['Day_of_Week'].isin(['Saturday','Sunday'])
data.head(5)
```

```python
data['Is_Holiday_Month']=data['Is_Holiday_Month'].map({'Yes':1, 'No':0})
data['Is_Holiday_Month']=data['Is_Holiday_Month'].astype(int)
```
